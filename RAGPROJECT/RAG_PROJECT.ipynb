```python
from pinecone import Pinecone, ServerlessSpec
```

```python
from google.colab import userdata
pc = Pinecone(api_key = userdata.get('srikar'))
```

```python
# What platform you are working with
cloud = "aws"
# What region you are working with
region = "us-east-1"
# Start the serverless database
serv = ServerlessSpec(cloud = cloud, region = region)
```

```python
# PLase dont use any capital letters, or underscore. Lowercase allowed with dash(-)
index_name = "srikar"
```

```python
# We need to check if there is a similar existing database or not
if index_name not in pc.list_indexes().names():
  # We will create a new index for ourselves
  pc.create_index(
      name = index_name,
      dimension = 768,
      metric = "cosine",
      spec = serv
  )

  # We will wait till the time our DB gets created and running
  while not pc.describe_index(index_name).status["ready"]:
    time.sleep(1)

print(pc.Index(index_name).describe_index_stats())
```

```python
from langchain.vectorstores import Pinecone
```

```python
# Here, we are check if the index exists or not
if index_name not in pc.list_indexes():
  # IF not, we will directly create and feed the data from the document
  docsearch = Pinecone.from_documents(docs, embedding, index_name = index_name)
else:
  # We will take the index and embedding model and then feed the data
  docsearch = Pinecone.from_existing_index(index_name, embedding, pinecone_index = pc.Index(index_name))
```

```python
docsearch
```

```python
repo_id = "mistralai/Mixtral-8x7B-Instruct-v0.1"
```

```python
from langchain.llms import HuggingFaceHub
```

```python
llm = HuggingFaceHub(
    repo_id = repo_id,   # Models
    huggingfacehub_api_token = userdata.get("HFToken") # Authentication Key (API Key)
)
```

```python
from langchain import PromptTemplate # Corrected the typo from PromtTemplate to PromptTemplate
```

```python
template = """
You are a MLOPs Engineer, the user is going to ask some questions about Machine Learning Operations.
Use the following context to answer the question.
IF YOU DON'T KNOW THE ANSWER, JUST SAY DON'T KNOW
KEEP THE ANSWER BRIEF

Context: {context}
Question: {question}
Answer:

"""
```

```python
prompt = PromptTemplate(
    template = template,
    input_variables = ["context", "question"]
)
```

```python
prompt
```

```python
from langchain.chains import RetrievalQA
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type = "stuff",
    retriever = docsearch.as_retriever()
)
```

```python
query = "What is MLOPs?"
```

```python
results = qa_chain.run(query)
```

```python
print(results)
```
